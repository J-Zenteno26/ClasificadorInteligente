## L2 - Deep Learning

### Arquitectura óptima para clasificación de imágenes

Para problemas de clasificación de imágenes, la arquitectura más adecuada es la **Red Neuronal Convolutiva (CNN)**. A diferencia de las redes densas, las CNN están diseñadas específicamente para trabajar con datos espaciales como imágenes, permitiendo detectar patrones locales como bordes, texturas y formas.

Una CNN típica está compuesta por:

* **Capas convolucionales (Conv2D)**: extraen características relevantes de la imagen.
* **Capas de pooling (MaxPooling)**: reducen la dimensionalidad manteniendo la información importante.
* **Capas densas (Dense)**: realizan la clasificación final a partir de las características extraídas.

Esta arquitectura permite capturar la estructura espacial de las imágenes, lo que resulta fundamental para lograr un buen desempeño en tareas de visión computacional.

---

### Justificación: red densa vs red convolutiva

Las redes neuronales densas, como la utilizada en la lección anterior, tratan cada píxel de la imagen como una entrada independiente, perdiendo la información espacial. Esto limita su capacidad para reconocer patrones visuales complejos.

En cambio, las redes convolutivas:

* preservan la estructura espacial de la imagen,
* reducen la cantidad de parámetros,
* detectan patrones locales de manera eficiente,
* generalizan mejor en tareas de visión.

Por estas razones, aunque las redes densas son útiles como punto de partida, las CNN son la opción óptima para clasificación de imágenes.

### Selección del framework: Keras

Para el desarrollo del proyecto se utilizará **Keras**, integrado en TensorFlow.

La elección se justifica por:

* su facilidad de uso y sintaxis clara,
* rápida implementación de modelos,
* amplia documentación y comunidad,
* integración directa con TensorFlow, lo que permite escalar modelos fácilmente.

Keras permite construir prototipos de forma eficiente, lo que lo convierte en una excelente opción para proyectos académicos y aplicaciones reales.
